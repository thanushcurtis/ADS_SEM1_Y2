```python
class Stack:
    def __init__(self):
        self.inList = LinkedList()

    def __str__(self):
        return str(self.inList)
        
    def size(self):
        return self.inList.length

    def push(self, e):
        self.inList.insert(0,e)

    def pop(self):
        return self.inList.remove(0)
    
s = Stack()
s.push(7); s.push(4); s.push(3)
print(s)
print(s.pop())
print(s)
s.push(42)
print(s)
```

```python
class Queue:
    def __init__(self):
        self.inList = LinkedList()

    def __str__(self):
        return str(self.inList)
        
    def size(self):
        return self.inList.length

    def enq(self, e):
        self.inList.append(e)

    def deq(self):
        return self.inList.remove(0)
    
q = Queue()
q.enq(7); q.enq(4); q.enq(3)
print(q)
print(q.deq())
print(q)
q.enq(42)
print(q)
```

```python
class Node:
    def __init__(self, d, n):
        self.data = d
        self.next = n

class LinkedList2:
    def __init__(self):
        self.head = self.tail = None
        self.length = 0

    def __str__(self):
        if self.head == None:
            return "empty"
        st = "-->"
        ptr = self.head
        while ptr != self.tail:
            st += " "+str(ptr.data)+" ->"
            ptr = ptr.next
        return st+" "+str(self.tail.data)+" ]"
        
    def search(self, d):
        i = 0
        ptr = self.head
        while ptr != None:
            if ptr.data == d:
                return i
            ptr = ptr.next
            i += 1
        return -1
        
    def append(self, d):
        if self.head == None:      
            self.head = self.tail = Node(d,None) 
        else:
            self.tail.next = Node(d,None)
            self.tail = self.tail.next
        self.length += 1

    def insert(self, i, d):
        if self.head == None or i == 0:
            self.head = Node(d,self.head)
            if self.tail == None: 
                self.tail = self.head
        else:
            ptr = self.head
            while i>1 and ptr.next != None:
                ptr = ptr.next
                i -= 1
            ptr.next = Node(d,ptr.next)
        if self.tail.next != None: 
            self.tail = self.tail.next
        self.length += 1

    def remove(self, i): # removes i-th element and returns it
        if self.head == None:
            return None
        if i == 0:
            val = self.head.data
            self.head = self.head.next
            self.length -= 1
            if self.head == None: # if the list is now empty
                self.tail = None
            return val
        ptr = self.head
        while ptr.next != None:
            if i == 1:
                val = ptr.next.data
                ptr.next = ptr.next.next
                self.length -= 1
                if ptr.next == None: # if we removed the tail
                    self.tail = ptr
                return val                
            ptr = ptr.next
            i -= 1
    
    def get(self, i):
        ptr = self.head
        while i > 0:
            ptr = ptr.next
            i -= 1
        return ptr.data
    
    def set(self, i, d):
        ptr = self.head
        while i > 0:
            ptr = ptr.next
            i -= 1
        ptr.data = d

# some testing
ls = LinkedList2()
print(ls,ls.length)
for x in [0,1,3]: ls.append(x)
print(ls,ls.head.data,ls.tail.data)
ls.insert(0,-2); ls.insert(1,-1)
print(ls,ls.head.data,ls.tail.data)
ls.insert(4,2)
print(ls,ls.head.data,ls.tail.data)
ls.insert(6,4)
print(ls,ls.head.data,ls.tail.data)

print(ls.search(1),ls.search(2))

print(ls.remove(0),ls,ls.head.data,ls.tail.data)
print(ls.remove(1),ls,ls.head.data,ls.tail.data)
print(ls.remove(4),ls,ls.head.data,ls.tail.data)
for _ in range(3): ls.remove(ls.length-1)
print(ls,ls.head.data,ls.tail.data)
print(ls.remove(0),ls)
```

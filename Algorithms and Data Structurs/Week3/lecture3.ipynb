```python
# search function with auxiliary search function
# - relies on indexing

def searchRec(A, k):
    return searchRecAux(A,k,0)

def searchRecAux(A, k, lo): # search for k in A[lo:]
    if lo == len(A): return -1
    if A[lo] == k: return lo
    return searchRecAux(A, k, lo+1)




print(searchRec([1,2,3,4,5,6],5))
print(searchRec([1,2,3,4,5,6,7],5))
print(searchRec([1,2,3,4,5,6,7],7))
print(searchRec([1,2,3,4,5,6,7],1))
print(searchRec([1,2,3,4,5,6],6))
print(searchRec([1,2,3,4,5,6],1))
print(searchRec([1,2,3,4,5,6],0))
print(searchRec([1,2,3,4,5,6,7],0))  
print(searchRec([],1))
```

```python
# recursive search function with sub
# - relies on subarray creation

def searchRec2(A, k):
    if A == []:
        return -1
    if A[0] == k:
        return 0
    recS = searchRec2(A[1:], k)
    if recS == -1:
        return -1
    return 1+recS

print(searchRec2([1,2,3,4,5,6],5))
print(searchRec2([1,2,3,4,5,6,7],5))
print(searchRec2([1,2,3,4,5,6,7],7))
print(searchRec2([1,2,3,4,5,6,7],1))
print(searchRec2([1,2,3,4,5,6],6))
print(searchRec2([1,2,3,4,5,6],1))
print(searchRec2([1,2,3,4,5,6],0))
print(searchRec2([1,2,3,4,5,6,7],0))  
print(searchRec2([],1))
```

```python
# 3 different ways to code factorial

# the iterative way
def fact(n):
    acc = 1
    for i in range(1,n+1):
        acc *= i
    return acc

# the recursive way
def factRec(n):
    if n == 0:
        return 1
    return n*factRec(n-1)

# the tail recursive way
def factRecTail(n):
    return factRecAux(n,1)

def factRecAux(n,acc):
    if n == 0:
        return acc
    return factRecAux(n-1,n*acc)

# now test them
import time

# put all the testing routine together in a function
def foo(factToUse, name, n):
    t=time.time()
    res=factToUse(n)
    t=time.time()-t
    print("- "+name+":\n  "+str(res)+"\n  in "+str(t)+" secs")

for n in [5,10,50,100,500]:
    print("\nfor n = "+str(n))
    foo(fact,"fact",n)
    foo(factRec,"factRec",n)
    foo(factRecTail,"factRecTail",n)

# now test with n = 1000
# depending on the local config, the last two may throw max recursion error

n=1000
foo(fact,"fact",n)
foo(factRec,"factRec",n)
foo(factRecTail,"factRecTail",n)
```

```python
def mergeSort(A):
    if len(A) <= 1: return 
    mid = len(A)//2

    half1 = A[:mid] # copy of A from position 0 up to mid-1
    half2 = A[mid:] # copy of A from position mid up to end
    mergeSort(half1) # recursion
    mergeSort(half2) # recursion
    merge(half1,half2,A)
    
def merge(half1, half2, A):
    i=0; j1=0; j2=0 # i: A, j1: half1, j2: half2
    while j1<len(half1) and j2<len(half2):
        if half1[j1] < half2[j2]:
            A[i] = half1[j1]
            j1 += 1
        else:
            A[i] = half2[j2]
            j2 += 1
        i += 1
    while j1 < len(half1):
        A[i] = half1[j1]
        j1 += 1; i += 1
    while j2 < len(half2):
        A[i] = half2[j2]
        j2 += 1; i += 1
            
A = [2,1,4,3]   
mergeSort(A)
print(A)
A = [3,21,65,1,3,1,3,4,5,11,5,656,4,5,6,7,12]
mergeSort(A)
print(A)
A = []
mergeSort(A)
print(A)
A = [1,2,3]
mergeSort(A)
print(A)
```

```python
# implementation with Θ(n) additional space

def mergeSort(A):
    B = A[:] # to be used as temporary storage
    mergeSortRec(A,0,len(A),B)

def mergeSortRec(A,lo,hi,B):
    if hi-lo <= 1: return 
    mid = (lo+hi)//2
    mergeSortRec(A,lo,mid,B)
    mergeSortRec(A,mid,hi,B)
    merge(A,lo,mid,hi,B)
    
def merge(A,lo, mid, hi, B):
    for i in range(lo,hi): 
        B[i] = A[i]
    i=lo; j1=lo; j2=mid 
    while j1<mid and j2<hi:
        if B[j1] < B[j2]:
            A[i] = B[j1]
            j1 += 1; i += 1
        else:
            A[i] = B[j2]
            j2 += 1; i += 1
    while j1 < mid:
        A[i] = B[j1]
        j1 += 1; i += 1
    while j2 < hi:
        A[i] = B[j2]
        j2 += 1; i += 1
            
A = [2,1,4,3]    
mergeSort(A)
print(A)
A = [3,21,65,1,3,1,3,4,5,11,5,656,5,1,1,5,1,41,1,4,5,6,7,12]
mergeSort(A)
print(A)
A = []
mergeSort(A)
print(A)
A = [1,2,3]
mergeSort(A)
print(A)
```

```python
def quickSort(A):
    quickSortRec(A,0,len(A))

def quickSortRec(A, lo, hi): # sort A from lo to hi-1    
    if hi-lo <= 1:
        return
    iPivot = partition(A,lo,hi)
    quickSortRec(A,lo,iPivot)
    quickSortRec(A,iPivot+1,hi)
    
def partition(A, lo, hi): # partition A from lo to hi-1
    pivot = A[lo]
    B = [0 for i in range(lo,hi)]
    loB = 0
    hiB = len(B)-1
    for i in range(lo+1,hi):
        if A[i] < pivot:
            B[loB] = A[i]
            loB += 1
        else:
            B[hiB] = A[i]
            hiB -= 1
    B[loB] = pivot
    for i in range(len(B)):
        A[lo+i] = B[i]
    return lo+loB

A = [2,1,4,3]    
quickSort(A)
print(A)
A = [3,21,65,1,3,1,3,4,5,11,5,656,5,1,1,5,1,41,1,4,5,6,7,12]
quickSort(A)
print(A)
A = []
quickSort(A)
print(A)
A = [1,2,3]
quickSort(A)
print(A)
```

```python
# implementation with Θ(1) additional space

def quickSort(A):
    quickSortRec(A,0,len(A))

def quickSortRec(A, lo, hi):    
    if hi-lo <= 1:
        return
    iPivot = partition(A,lo,hi)
    quickSortRec(A,lo,iPivot)
    quickSortRec(A,iPivot+1,hi)
    
# in-place variant
def partition(A, lo, hi):
    pivot = A[lo]
    loB = lo
    hiB = hi-1
    i = lo+1
    while i<hiB:
        if A[i] < pivot:
            A[loB] = A[i]
            loB += 1
            i += 1
        else:
            swap(hiB,i,A)
            hiB -= 1
    A[loB] = pivot
    return loB

def swap(i,j,A):
    tmp = A[i]
    A[i] = A[j]
    A[j] = tmp

A = [2,1,4,3]    
quickSort(A)
print(A)
A = [3,21,65,1,3,1,3,4,5,11,5,656,5,1,1,5,1,41,1,4,5,6,7,12]
quickSort(A)
print(A)
A = []
quickSort(A)
print(A)
A = [1,2,3]
quickSort(A)
print(A)
```

```python
class Script:
    def __init__(self, sid, mark):
        self.sid = sid
        self.mark = mark
        
    # in order to print arrays of Scripts, we need to define the repr function
    def __repr__(self):
        return "("+str(self.sid)+","+str(self.mark)+")"
        
def quickSort(A):
    quickSortRec(A,0,len(A))

def quickSortRec(A, lo, hi):    
    if hi-lo <= 1:
        return
    iPivot = partition(A,lo,hi)
    quickSortRec(A,lo,iPivot)
    quickSortRec(A,iPivot+1,hi)
    
def partition(A, lo, hi):
    pivot = A[lo]
    B = [None for i in range(lo,hi)]
    loB = 0
    hiB = len(B)-1
    for i in range(lo+1,hi):
        if A[i].sid < pivot.sid:
            B[loB] = A[i]
            loB += 1
        else:
            B[hiB] = A[i]
            hiB -= 1
    B[loB] = pivot
    for i in range(lo,hi):
        A[i] = B[i-lo]
    return lo+loB

A = [None for i in range(4)]
A[0] = Script(133,59)
A[1] = Script(130,54)
A[2] = Script(143,80)
A[3] = Script(163,43)

quickSort(A)
print(A)
```

```python
# a maze class

class Maze:
    def __init__(self, width, string):
        self.width = width
        self.height=len(string)//width
        self.position = [[False for i in range(width)] for j in range(self.height)]
        for p in range(len(string)) :
            h=p//width
            w=p%width
            self.position[self.height-1-h][w] = string[p]==' '
        for i in range(self.width):
            if self.position[0][i]:
                self.enter = i
            if self.position[self.height-1][i]:
                self.exit = i
    
    def __str__(self):
        string='\n'
        for i in range(self.height):
            line=self.position[self.height-1-i]
            for j in range(len(line)) :
                if line[j] :
                    string=string+' '
                else :
                    string=string+'X'
            string=string+'\n'
        return string

    
mazestring1='\
XXXX XXXX\
X     XXX\
X X X   X\
X XXXXX X\
X       X\
XXXXX XXX\
'

mazestring2='\
XXXXXXXXXX X\
X  XXX X X X\
X XX   X   X\
X  X XXX XXX\
X XX X     X\
X    X XXX X\
XXXX X X   X\
XXXX X X XXX\
X    X X   X\
X XXXX X XXX\
X        XXX\
XXXXX XXXXXX\
'
    
maze1 = Maze(9,mazestring1)
maze2 = Maze(12,mazestring2)
print(maze1)
print(maze2)
```

```python
class ArrayList:
    def __init__(self):  #  we need an internal array and a count
        self.inArray = [0 for i in range(10)]
        self.count = 0
        
    def get(self, i):  # we do not check bounds!
        return self.inArray[i]

    def set(self, i, e):  # we do not check bounds!
        self.inArray[i] = e

    def length(self):
        return self.count

    def append(self, e):  # assumption: there is always at least one empty space to append
        self.inArray[self.count] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()     # resize array if reached capacity         
            
    def insert(self, i, e):  # no bound checks!
        for j in range(self.count,i,-1):
            self.inArray[j] = self.inArray[j-1]
        self.inArray[i] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()     # resize array if reached capacity

    def remove(self, i):
        self.count -= 1
        val = self.inArray[i]
        for j in range(i,self.count):
            self.inArray[j] = self.inArray[j+1]
        return val
    
    def __str__(self):
        if self.count == 0: return "[]"
        s = "["
        for i in range(self.count-1): s += str(self.inArray[i])+", "
        return s+str(self.inArray[self.count-1])+"]"

    def _resizeUp(self):  # is called when len(inArray) == count
        newArray = [0 for i in range(2*len(self.inArray))]
        for j in range(len(self.inArray)):
            newArray[j] = self.inArray[j]
        self.inArray = newArray


# some testing
ls = ArrayList()
for i in range(9): ls.append(i)
print(ls,"= {",ls.inArray,",", ls.count,"}")

ls.append(9)
print(ls,"= {",ls.inArray,",", ls.count,"}")

ls.set(2,24)
print(ls,"= {",ls.inArray,",", ls.count,"}")

print(ls.get(0), ls.length())
ls.insert(4,42)
print(ls,"= {",ls.inArray,",", ls.count,"}")

print(ls.remove(4))
print(ls,"= {",ls.inArray,",", ls.count,"}")
```

```python
class Queue:
    
    def __init__(self):
        self.inList = ArrayList()
        self.tail = 0  # the queue is from tail to count-1

    def size(self):
        return self.inList.length()-self.tail

    def enq(self,e): # Theta(1) amortized
        self.inList.append(e)

    def deq(self): # Theta(1) 
        val = self.inList.get(self.tail)
        self.tail += 1
        return val
    
    def enqAll(self,A):
        for x in A:
            self.enq(x)
            
    def __str__(self): # without using internal fields of ArrayList
        if self.size()==0: return "[]"
        s = "["
        for i in range(self.tail,self.inList.length()-1):
            s += str(self.inList.get(i))+","
        return s+str(self.inList.get(self.inList.length()-1))+"]"

q = Queue()
q.enqAll([2,3,4,5,42])
print(q)
q.enq(3); print(q)
q.deq(); print(q)
q.enq(23); print(q)

for i in range(6):
    print(q.deq(),q)
```
